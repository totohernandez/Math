# Modelos Causales

## Regresion Lineal Simple

El **analisis de regresion** nos permite obtener una ecuacion que nos idica cual es la relacion entre las variables. 

La **variable dependiente** es la variable que se va predecir. 

Las **variables independientes** se usan para predecir el valor de la variable dependiente. 

La **regresion lineal simple** es el tipo mas sencillo de analisis de regresion, en el cual interviene una variable independiente y una variable dependiente, y la relacion entre estas variables es aproximadamente una linea recta. 

La **regresion multiple** es el analisis en el cual intervienen dos o mas variables dependientes. 

El **modelo de regresion** es la ecuacion con la cual se describe como se relacionan *x* con *y*, y se da un termino para el error. 

### Modelo de Regresion Lineal Simple. 

$ y = \beta_0 + \beta_1 x + \epsilon $

- $ \beta_0 $ y $ \beta_1 $ se conocen como los parametros del modelo. 

- $ \epsilon $ es una variable aleatoria que se conoce como termino del error. 

- El termino del error da cuenta de la variabilidad de *y* y que no puede ser explicada por la relacion lineal entre *x* y *y*.

### Ecuacion de Regresion Lineal Simple

La **ecuacion de regresion** describe la relacion entre el valor esperado *y* (que se denota $ E(y) $), y *x*. 

$ E(y) = \beta_0 + \beta_1 x $

- La grafica de la ecuacion de regresion lineal es una linea recta. 
- $ \beta_0 $ es la interseccion de la recta de regresion con el eje *y*, $ \beta_1 $ es la pendiente y $ E(y) $ es la media o valor esperado de *y* para un valor dado de *x*.

### Estimacion de la Regresion Lineal Simple

Si no conocemos los valores de los parametros poblacionales $ \beta_0 $ y $ \beta_1 $, es necesario estimarlos usando datos muestrales. 

Se calculan estadisticos muestrales (que se denotan $b_0$ y $b_1$) como estimaciones de los parametros poblacionales. 

La ecuacion de regresion estimada: 

$ \hat{y} = b_0 + b_1 x $

### Coeficiente de Determinacion 

El **coeficiente de determinacion** es una medida de la bondad de ajuste de la ecuacion estimada, es decir lo bien que se ajusta la ecuacion a los datos. 

#### Coficiente de Determinacion

$ r^2 = \frac{SRC}{STC} $

#### Suma Total de Cuadrados

$ STC = \sum (y_i - \bar{y})^2 $

#### Suma de Cuadrados Debida a la Regresion

$ SCR = \sum (\hat{y_i} - \bar{y})^2 $

### Suposiciones del Modelo

1. El termino del error $ \epsilon $ es una variable aletoria cuya media, o valor esperado, es cero: es decir, $ E(\epsilon) = 0 $

2. La varianza de $ \epsilon $, que se denota $ \sigma^2 $, es la misma para todos los valores de *x*. 

3. Los valores de $ \epsilon $ son independientes. 

4. El termino del error $ \epsilon $ son independientes. 

### Formulacion Estimadores con Minimos Cuadrados

$ Min (\sum_{i=1}^{n} d_i) $

$ Min (\sum_{i=1}^{n} (y_i - \hat{y_i})^2) $

$ \hat{y_i} = \hat{\beta_0} + \hat{\beta_1} x_i $

$ Min (\sum_{i=1}^{n} (y_i - (\hat{\beta_0} + \hat{\beta_1} x_i))^2) $

#### Suma de Cuadrados Debido a los Residuos

$ SSR = \sum (y_i - \hat{y_i})^2 $

#### Minimo usando Calculo

$ SSR = \sum_{i=1}^{n} (y_i - (\hat{\beta_0} + \hat{\beta_1} x_i))^2 $

$ \frac{\partial SSR}{\partial \hat{\beta}} \rightarrow \frac{\partial SSR}{\partial \hat{\beta_0}} = 0, \frac{\partial SSR}{\partial \hat{\beta_1}} = 0 $

Derivando respecto a $ \hat{\beta_0} $
 
$ \frac{\partial SSR}{\partial \hat{\beta_0}} = \frac{\partial}{\partial \hat{\beta_0}} \sum_{i=1}^{n} (y_i - (\hat{\beta_0} + \hat{\beta_1} x_i))^2 $

$ = \sum_{i=1}^{n} \frac{\partial}{\partial \hat{\beta_0}} (y_i - (\hat{\beta_0} + \hat{\beta_1} x_i))^2$

$ = 2 * \sum_{i=1}^{n} \frac{\partial}{\partial \hat{\beta_0}} (y_i - (\hat{\beta_0} + \hat{\beta_1} x_i))^1 *  [(\frac {\partial}{\partial \hat{\beta_0}} y_i = 0) - (\frac {\partial}{\partial \hat{\beta_0}} \hat{\beta_0} - \hat{\beta_1} x_i = 1)] $

$ = 2 * \sum_{i=1}^{n} \frac{\partial}{\partial \hat{\beta_0}} (y_i - (\hat{\beta_0} + \hat{\beta_1} x_i)) * (-1)$

$ = -2 * \sum_{i=1}^{n} \frac{\partial}{\partial \hat{\beta_0}} (y_i - (\hat{\beta_0} + \hat{\beta_1} x_i)) \rightarrow Ecuacion 1 $

Derivando respecto a $ \hat{\beta_1} $

$ \frac{\partial SSR}{\partial \hat{\beta_1}} = \frac{\partial}{\partial \hat{\beta_1}} \sum_{i=1}^{n} (y_i - (\hat{\beta_0} + \hat{\beta_1} x_i))^2 $

$ = \sum_{i=1}^{n} \frac{\partial}{\partial \hat{\beta_1}} (y_i - (\hat{\beta_0} + \hat{\beta_1} x_i))^2$

$ = 2 * \sum_{i=1}^{n} \frac{\partial}{\partial \hat{\beta_1}} (y_i - (\hat{\beta_0} + \hat{\beta_1} x_i))^1 *  [(\frac {\partial}{\partial \hat{\beta_1}} y_i = 0) - (\frac {\partial}{\partial \hat{\beta_1}} \hat{\beta_0} - \hat{\beta_1} x_i = x_i)] $

$ = 2 * \sum_{i=1}^{n} \frac{\partial}{\partial \hat{\beta_1}} (y_i - (\hat{\beta_0} + \hat{\beta_1} x_i)) * (-x_i)$

$ = -2 * \sum_{i=1}^{n} \frac{\partial}{\partial \hat{\beta_1}} (y_i - (\hat{\beta_0} + \hat{\beta_1} x_i)) * x_i \rightarrow Ecuacion 2 $

Igualando las ecuaciones a zero

$ (1) -2 * \sum_{i=1}^{n} \frac{\partial}{\partial \hat{\beta_0}} (y_i - (\hat{\beta_0} + \hat{\beta_1} x_i)) = 0 $

$ (2) -2 * \sum_{i=1}^{n} \frac{\partial}{\partial \hat{\beta_1}} (y_i - (\hat{\beta_0} + \hat{\beta_1} x_i)) * x_i = 0 $

Solucion para $\hat{\beta_0}$ :

$ (1) -2 * \sum_{i=1}^{n} \frac{\partial}{\partial \hat{\beta_0}} (y_i - (\hat{\beta_0} + \hat{\beta_1} x_i)) = 0 $

$ \sum_{i=0}^{n} y_i - \hat{\beta_0} \sum_{i=0}^{n} 1 - \hat{\beta_1} \sum_{i=1}^{n} x_i = 0 $

$ \sum_{i=0}^{n} y_i - \hat{\beta_0} * n - \hat{\beta_1} \sum_{i=1}^{n} x_i = 0 $

$ \sum_{i=0}^{n} y_i - \hat{\beta_1} \sum_{i=1}^{n} x_i = \hat{\beta_0} * n $

$ \frac{\sum_{i=0}^{n} y_i}{n} - \hat{\beta_1} \frac{\sum_{i=1}^{n} x_i}{n} = \hat{\beta_0} $

Concluimos

$ \hat{\beta_0} = \bar{y} - \hat{\beta_1} * \bar{x} $

Solucion para $\hat{\beta_1}$ :

$ (2) -2 * \sum_{i=1}^{n} \frac{\partial}{\partial \hat{\beta_1}} (y_i - (\hat{\beta_0} + \hat{\beta_1} x_i)) * x_i = 0 $

$ \sum_{i=1}^{n} x_i * (y_i - ((\bar{y} - \hat{\beta_1} * \bar{x}) + (\hat{\beta_1} x_i)) = 0 $ 

$ \sum_{i=1}^{n} x_i * (y_i - \bar{y} + \hat{\beta_1} * \bar{x} - \hat{\beta_1} x_i) = 0 $ 

$ \sum_{i=1}^{n} x_i * (y_i - \bar{y} - \hat{\beta_1} * (x_i -\bar{x})) = 0 $ 

$ \sum_{i=1}^{n} x_i * (y_i - \bar{y}) - \hat{\beta_1} * x_i *(x_i -\bar{x}) = 0 $ 

$ \sum_{i=1}^{n} x_i * (y_i - \bar{y}) - \hat{\beta_1} * \sum_{i=1}^{n} x_i *(x_i -\bar{x}) = 0 $

$ \sum_{i=1}^{n} x_i * (y_i - \bar{y}) = \hat{\beta_1} * \sum_{i=1}^{n} x_i *(x_i -\bar{x}) $

Concluimos

$ \hat{\beta_1} = \frac{\sum_{i=1}^{n} x_i * (y_i - \bar{y})}{\sum_{i=1}^{n} x_i *(x_i -\bar{x})}$

Equivalente a:

$ \hat{\beta_1} = \frac{\sum_{i=1}^{n} (x_i - \bar{x})(y_i - \bar{y})}{\sum_{i=1}^{n} (x_i - \bar{x})^2} $
